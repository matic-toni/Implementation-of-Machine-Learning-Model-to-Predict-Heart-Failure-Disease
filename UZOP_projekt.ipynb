## Strojno učenje
Neki koraci iz rada koji su sređivali podatke su preskočeni s razlogom. Na primjer, procijenio sam da nije potrebno raditi augmentaciju podataka, s obzirom na to da je autor rada imao bazu s 300 primjera koju je povećao na oko 900. Budući da ja koristim već proširenu bazu podataka od 918 primjera, ovaj sam korak preskočio.

Dakle, nakon što smo gotovi s pripremom podataka, možemo nastaviti na drugi korak rada - strojno učenje. Pozabavit ćemo se algoritmima koji će obavljati predikciju novodobivenih primjera u budućnosti.

Koristit ćemo 5 algoritama, a to su redom:
- Stablo odluke (Decision Tree)
- Logistička regresija (Logistic Regression)
- Slučajna šuma (Random Forest)
- Naivni Bayesov klasifikator (Naive Bayes)
- Stroj potpornih vektora (SVM)

Prvo ćemo napraviti import potrebnih biblioteka

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, plot_confusion_matrix
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import CategoricalNB
from sklearn.svm import SVC
```

Budući da ovi algoritmi ne rade s kategorijskim varijablama koja sadrže nenumeričke vrijednosti, prvo ćemo njih zamijeniti s klasama koje su određene brojevima.

```python
heart['Sex'].replace(to_replace='F', value=0, inplace=True)
heart['Sex'].replace(to_replace='M', value=1, inplace=True)

heart['ChestPainType'].replace(to_replace='TA', value=0, inplace=True)
heart['ChestPainType'].replace(to_replace='ATA', value=1, inplace=True)
heart['ChestPainType'].replace(to_replace='NAP', value=2, inplace=True)
heart['ChestPainType'].replace(to_replace='ASY', value=3, inplace=True)

heart['RestingECG'].replace(to_replace='Normal', value=0, inplace=True)
heart['RestingECG'].replace(to_replace='ST', value=1, inplace=True)
heart['RestingECG'].replace(to_replace='LVH', value=2, inplace=True)

heart['ExerciseAngina'].replace(to_replace='N', value=0, inplace=True)
heart['ExerciseAngina'].replace(to_replace='Y', value=1, inplace=True)

heart['ST_Slope'].replace(to_replace='Up', value=0, inplace=True)
heart['ST_Slope'].replace(to_replace='Flat', value=1, inplace=True)
heart['ST_Slope'].replace(to_replace='Down', value=2, inplace=True)
```

## Stablo odluke (Decision Tree)
Ovo je klasifikacijski algoritam koji ćemo prvi iskoristiti za učenje našeg modela poznatim, ali i predikciju klasa za buduće primjere. Budući da ovaj algoritam ne radi s kontinuiranim varijablama, prvo ćemo vrijednosti takvih značajki mapirati u neke određene grupe. To sam napravio tako da sam u nekim literaturama koje se bave zdravstvenim problematikama pronašao neke granice normalnih vrijednosti za određenu značajku te sam te brojeve prema njima razvrstao u kategorije. Svaki je broj upao ili u klasu 0 (zdrava, normalna vrijednost) ili 1 (nezdrava vrijednost značajke).

Nakon što sam to napravio, podijelio sam pripremljeni skup podataka na dva dijela - dio za učenje i dio za testiranje. Tu podjelu sam napravio u omjeru 70:30. Nakon toga, stvorio sam model Stabla odluke koji sam i istrenirao na skupu za učenje. Na X skupu za testiranje napravio sam predikciju izlaznih (y) vrijednosti, te sam proveo provjeru ispravnosti rada modela. Zbog točnosti, kao i u izvornom radu, ovaj sam ovaj postupak ponovio 5 puta, te je konačan rezultat uzet kao prosjek dobivenih vrijednosti.

Na takav način, dobiven je prosječan rezultat točnosti modela koji iznosi nešto više od 82%.
Osim toga, nacrtana je matrica konfuzije za 1 iteraciju algoritma

```python
X = heart.loc[:, heart.columns != 'HeartDisease'].copy()
y = heart['HeartDisease'].copy()

X.loc[:, 'Age'] = pd.cut(heart.Age, bins=[28, 50, 77], labels=[0, 1], include_lowest=True)
X.loc[:, 'RestingBP'] = pd.cut(heart['RestingBP'], bins= [80, 120, 200], labels=[0, 1], include_lowest=True)
X.loc[:, 'Cholesterol'] = pd.cut(heart['Cholesterol'], bins= [1, 200, 605], labels=[0, 1], include_lowest=True)
X.loc[:, 'MaxHR'] = pd.cut(heart['MaxHR'], bins= [71, 100, 202], labels=[0, 1], include_lowest=True)
X.loc[:, 'Oldpeak'] = pd.cut(heart['Oldpeak'], bins= [0, 1, 6.2], labels=[0, 1], include_lowest=True)

res = []

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

    clf = DecisionTreeClassifier()
    clf = clf.fit(X_train, y_train)

    y_predicted = clf.predict(X_test)
    res.append(accuracy_score(y_test, y_predicted))
        
plot_confusion_matrix(clf, X_test, y_test)

print('Ispravnost u iteracijama:', res)
print('Prosječna vrijednost ispravnosti za 5 iteracija je:', mean(res))
```

## Logistička regresija (Logistic Regression)
Logistička je regresija algoritam koji klase odvaja u n-dimenzionalnom prostoru pomoću neke hiperravnine koja određuje granicu između tih klasa. U dvodimenzijskom prostoru, kakvog imamo i u ovom primjeru, to je neki pravac. Ovdje, kao i kod stabla odluke, dijelimo skup podataka na skup za učenje i skup za testiranje, gradimo model, treniramo ga te ga koristimo za predikciju. Točnost ovog modela iznosi oko 87%.

```python
X = heart.loc[:, heart.columns != 'HeartDisease']
y = heart['HeartDisease']

res = []

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

    clf = LogisticRegression(random_state=0, max_iter=1000).fit(X_train, y_train)

    y_predicted = clf.predict(X_test)
    res.append(accuracy_score(y_test, y_predicted))

plot_confusion_matrix(clf, X_test, y_test)
        
print('Ispravnost u iteracijama:', res)
print('Prosječna vrijednost ispravnosti za 5 iteracija je:', mean(res))
```

## Slučajna šuma (Random Forest)
Slučajna je šuma algoritam koji koristi puno stabala odluke (u našem primjeru ih je 100) koji onda, svaki pojedinačno donosi odluku u koju klasu svrstati novi primjer. Konačna klasa se odabire na temelju glasanja, tj. pobjeđuje ona koja je najviše puta izabrana. Postupak je isti kao u prethodnim primjerima. Točnost klasifikacije ovim modelom iznosi oko 88%

```python
X = heart.loc[:, heart.columns != 'HeartDisease']
y = heart['HeartDisease']

res = []

for i in range(5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

    clf = RandomForestClassifier(n_estimators=100, random_state=0).fit(X_train, y_train)

    y_predicted = clf.predict(X_test)
    res.append(accuracy_score(y_test, y_predicted))

plot_confusion_matrix(clf, X_test, y_test)
    
print('Ispravnost u iteracijama:', res)
print('Prosječna vrijednost ispravnosti za 5 iteracija je:', mean(res))
```

## Naivni Bayesov klasifikator (Naive Bayes)
Ovaj je klasifikacijski algoritam ujedno i probabilistički. Za pojedini primjer računa vjerojatnost njegovog pripadanja svakoj mogućoj izlaznoj klasi. Na kraju, svrstava primjer u onu klasu u koju je najsigurniji da pripada.
Budući da implementacija ovog algoritma ponekad zna baciti iznimku pri predikciji, dodao sam try-except blok da se program pri računanju ne ruši. Prosječni rezultat je oko 87%.

```python
X = heart.loc[:, heart.columns != 'HeartDisease']
y = heart['HeartDisease']

 
res = []

while len(res) != 5:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

    try:
        clf = CategoricalNB().fit(X_train, y_train)
        y_predicted = clf.predict(X_test)
    except:
        pass

    res.append(accuracy_score(y_test, y_predicted))

plot_confusion_matrix(clf, X_test, y_test)
    
print('Ispravnost u iteracijama:', res)
print('Prosječna vrijednost ispravnosti za 5 iteracija je:', mean(res))
```

## Stroj potpornih vektora (SVM)
